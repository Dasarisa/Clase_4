```python
import math
from pylab import sin, sqrt
salida=open("ondas_est.txt","w")
x1=0
y1=0
x2=50
y2=50
landa=5.0
e0=1.0
k=2*math.pi/landa
nx=500
dimx=100
x=dimx/nx
ny=500
dimy=100
y=dimy/ny
for i in range(501):
    for j in range (501):
        r1=(math.sqrt(((x1-(i*x))**2)+((y1-(j*y))**2)))
        r2=(math.sqrt(((x2-(i*x))**2)+((y2-(j*y))**2)))
        e1=e0*(sin(k*r1))
        e2=e0*(sin(k*r2))
        e=e1+e2
        salida.write("%0.2f " %(e1+e2))
    salida.write("\n")
salida.close()
```
